```python
# 1. Make a scatterplot of the highway miles per gallon (y-axis) and weight (x-axis)

# import all requires libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("car_data.csv")
df
```

```python
get_ipython().magic('matplotlib inline')
plt.scatter(df.weight, df.hwy_mpg, color = "Green", marker = "+")
plt.xlabel("Vehicle Weight")
plt.ylabel("Miles per Gallon")
plt.title("Vehicles weight vs. MPG")
```

2. Based on the plot, what is the general trend of how highway miles per gallon varies with the weight?

    Based on the above scatter plot, it appears as though the heavier the weight of the vehicle, the lower the miles per gallon.
